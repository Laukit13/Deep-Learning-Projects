```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df.drop(['Cabin'], axis=1, inplace=True)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
print(df['Embarked'].mode())
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
df.head()
```

```python
X = df.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = df['Survived']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['Age'] = X_train_scaled['Age'] / 100
X_train_scaled['Fare'] = X_train_scaled['Fare'] / 100
X_train_scaled['Pclass'] = X_train_scaled['Pclass'] / 100
X_train_scaled['Parch'] = X_train_scaled['Parch'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['Age'] = X_test_scaled['Age'] / 100
X_test_scaled['Fare'] = X_test_scaled['Fare'] / 100
X_test_scaled['Pclass'] = X_test_scaled['Pclass'] / 100
X_test_scaled['Parch'] = X_test_scaled['Parch'] / 100
```

```python
print(X_test_scaled)
print(X_train_scaled)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(7,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')   
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_scaled, Y_train, epochs=1000) #Idle Epochs =1000 Accuracy = 80%
```

```python
model.evaluate(X_test_scaled,Y_test)
```
